HR-аналитика 
=

**Описание проекта:** HR-аналитики компании «Работа с заботой» помогают бизнесу оптимизировать управление персоналом: бизнес предоставляет данные, а аналитики предлагают, как избежать финансовых потерь и оттока сотрудников. В этом HR-аналитикам пригодится машинное обучение, с помощью которого получится быстрее и точнее отвечать на вопросы бизнеса.
Компания предоставила данные с характеристиками сотрудников компании. Среди них — уровень удовлетворённости сотрудника работой в компании. Эту информацию получили из форм обратной связи: сотрудники заполняют тест-опросник, и по его результатам рассчитывается доля их удовлетворённости от 0 до 1, где 0 — совершенно неудовлетворён, 1 — полностью удовлетворён.

**Цель проекта:** построить модель, которая сможет предсказать уровень удовлетворённости сотрудника на основе данных заказчика и построить модель, которая сможет на основе данных заказчика предсказать то, что сотрудник уволится из компании.


**Ход исследования:**
1. Подготовка данных - загрузка и изучение общей информации из предоставленных датасетов.
2. Предобработка данных - поиск пропущенных значений, дубликатов, замена названий столбцов (приведение к нижнему регистру), поиск аномальных значений.
3. Исследовательский анализ данных - изучение распределений количественных и категориальных признаков в датасетах, определение корреляции между признаками.
4. Обучение моделей регрессии для выбора наилучшей (с наименьшей метрикой SMAPE) для предсказания уровня удовлетворенности сотрудника.
5. Обучение моделей классификации для выбора наилучшей (с наибольшей метрикой ROC_AUC) для предсказания оттока сотрудников.

**Общий вывод:** Предложение лучших моделей МО для прогнозирования уровня удовлетворенности сотрудника и возможного оттока сотрудников.

## Предсказание уровня удовлетворенности сотрудника с помощью моделей МО

### Подготовка данных

Поскольку целевой признак представляет собой непрерывную числовую переменную, то стоит использовать модели линейной регресии и дерево решений для задач регресии.

```python
data_1 = data.drop('id', axis=1)

data_1.duplicated().sum()
```

```python
data_1 = data_1.drop_duplicates() 
```

```python
# создание тренировочной выборки входных признаков и целевого признака
X_train = data_1.drop('job_satisfaction_rate', axis=1)
y_train = data_1['job_satisfaction_rate']
X_test = X_test.drop('id', axis=1)
y_test = y_test['job_satisfaction_rate']
```

```python
#создание списков количественных и категориальных датасета data
num_col = ['employment_years', 
           'supervisor_evaluation', 
           'salary']

ord_col = ['level',
           'workload']

ohe_col = ['dept',
           'last_year_promo', 
           'last_year_violations']
```

```python
#Pipeline для кодирования категориальных признаков
ohe_pipe = Pipeline(
    [
        ('simpleImputer_ohe', 
         SimpleImputer(missing_values=np.nan, 
                       strategy='most_frequent')), 
        ('ohe', 
         OneHotEncoder(drop='first', 
                       handle_unknown='error', 
                       sparse=False))
    ]
)
```

```python
#Pipeline для кодирования категориальных ранговых признаков
ord_pipe = Pipeline(
    [('simpleImputer_before_ord', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
     ('ord',  OrdinalEncoder(
                categories=[
                    ['junior', 'middle', 'senior'], 
                    ['low', 'medium', 'high'],
                ], 
                handle_unknown='use_encoded_value', unknown_value=np.nan
            )
        ),
     ('simpleImputer_after_ord', SimpleImputer(missing_values=np.nan, strategy='most_frequent'))
    ]
)
```

```python
# Pipeline для подготовки данных
data_preprocessor = ColumnTransformer(
    [
        ('ohe', ohe_pipe, ohe_col),
        ('ord', ord_pipe, ord_col),
        ('num', StandardScaler(), num_col)
    ], 
    remainder='passthrough'
)
```

```python
# Итоговый Pipeline: подготовка данных и модель
pipe_final = Pipeline([
    ('preprocessor', data_preprocessor),
    ('models', DecisionTreeRegressor(random_state=RANDOM_STATE))
])
```

```python
#Параметры для передачи GridSearchCV для поиска оптимальной модели 
#с разными Scaler и Encoder и перебором гиперпараметров

param_distributions = [
    {
        'models': [DecisionTreeRegressor(random_state=RANDOM_STATE)],
        'models__max_depth': range(2, 21),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
    },
    {
        'models': [LinearRegression()],
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
    }, 
] 
```

### Обучение моделей

```python
# создание собственной метрики для оценки модели

def SMAPE(y_true, y_pred):
    return 100 / len(y_true) * np.sum(2 * np.abs(y_true - y_pred) / ((np.abs(y_true) + np.abs(y_pred))))
                                      
SMAPE_score = make_scorer(SMAPE, greater_is_better=False)
```

```python
#поиск оптимальной модели с перебором гиперпараметров
grid = GridSearchCV(
    pipe_final, 
    param_grid=param_distributions, 
    cv=5, 
    scoring=SMAPE_score, 
    n_jobs=-1, 
)

#обучение модели на тренировочной выборке
grid.fit(X_train, y_train);
```

```python
print('Метрика SMAPE для лучшей модели:\n', round(-grid.best_score_, 2)) 
print('\nЛучшая модель и её параметры:\n\n', grid.best_params_) 
```

```python
y_pred = grid.predict(X_test)
y_true = y_test
print(f'Метрика SMAPE на тестовой выборке: {round(SMAPE(y_true, y_pred), 2)}')
```

### Промежуточные выводы

```python
#поиск важных признаков
X_encoded = grid.best_estimator_['preprocessor'].fit_transform(X_train)

model = grid.best_estimator_['models'].fit(X_encoded, y_train)
explainer = shap.PermutationExplainer(model, X_encoded)


labels = grid.best_estimator_[:-1].get_feature_names_out()

plt.title('Оценка важности признаков')
plt.ylabel('Признаки')
shap.summary_plot(X_encoded, plot_type='bar', feature_names=labels) 
```

Лучшей оказалась модель дерева решений с глубиной равной 13, в которой в качестве масштабирования количественных признаков используется MinMaxScaler. 

На тестовой выборке метрика SMAPE улучшилась и получилась 14.43, что подходит к условию задачи, где критерием успеха является SMAPE <= 15.

Возможно эта модель оказалась лучше линейной регрессии, потому что возможно у нас есть нелинейные зависимости между входными и целевым признаками.

Наиболее важным признаком оказалсь нагрузка.

## Зависимость между job_satisfaction_rate и quit

```python
#добавление в тестовую выборку целевого признака из предыдущей задачи
X_test_1['job_satisfaction_rate'] = y_test
X_test_1['quit'] = y_test_1['quit']
X_test_1
```

Необходимо понять есть ли зависимость между уровнем удовлетворенности сотрудника и его желанием уволиться. Для этого проверим равны ли средние значения уровня удовлетворенности для каждой из групп. 

Сформулируем **нулевую гипотезу Н0**: средние уровени удовлетворенности уволившихся и оставшихся сотрудников равны.

**Альтернативная гипотеза** будет выглядеть следующим образом: средние уровени удовлетворенности уволившихся и оставшихся сотрудников не равны (двусторонняя гипотеза Н1).

Для проверки гипотезы о равенстве среднего двух генеральных совокупностей по взятым из них выборкам будем использовать метод st.ttest_ind(array1, array2, equal_var, alternative), поскольку данные выборки являются независимыми.

```python
hist_box_1(X_test_1, 'job_satisfaction_rate') #визуализация распределений
```

```python
#проверка гипотезы о равенстве средней продолжительности поездки для пользователей с подпиской и без
alpha = 0.05 #уровень значимости 

results = st.ttest_ind(X_test_1.query('quit == "no"')['job_satisfaction_rate'], 
                       X_test_1.query('quit == "yes"')['job_satisfaction_rate']) #расчет p_value
```

```python
print('Средний уровень удовлетворённости уволившегося сотрудника:', round(X_test_1.query('quit == "yes"')['job_satisfaction_rate'].mean(), 2))
print('Средний уровень удовлетворённости оставшегося сотрудника:', round(X_test_1.query('quit == "no"')['job_satisfaction_rate'].mean(), 2))
print('p-значение:', results.pvalue)

if results.pvalue < alpha: #проверка p_value
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

Отвергая нулевую гипотезу, мы действительно можем предположить, что у уволившихся сотрудников был меньший уровень удовлетворенности, чем у оставшихся.

Поэтому добавим признак уровня удовлетворенности во входные признаки для обучения новой модели.

### Подготовка данных

```python
df_1 = df.drop('id', axis=1) #создание новой тренировочной выборки
df_1.duplicated().sum()
```

```python
df_1 = df_1.drop_duplicates().reset_index(drop=True)
```

```python
X_train_1 = df_1.drop('quit', axis=1)

y = grid.predict(X_train_1) #предсказание уровня удовлетворенности для новой тестовой выборки

X_train_1['job_satisfaction_rate'] = pd.DataFrame(y) #добавление нового входного признака
X_train_1
```

```python
# создание тренировочной выборки входных признаков и целевого признака
y_train_1 = df_1['quit']
X_test_1 = X_test_1.drop(['id', 'quit'], axis=1)
y_test_1 = y_test_1['quit']
```

```python
#создание списков количественных и категориальных датасета df
num_col_1 = ['employment_years', 
           'supervisor_evaluation', 
           'salary', 
           'job_satisfaction_rate']

ohe_col_1 = ['dept',
           'level',
           'workload',
           'last_year_promo', 
           'last_year_violations']

ord_col_1 = ['level',
             'workload']
```

```python
#кодирование целевого признака
label_encoder = LabelEncoder()
y_train_1 = label_encoder.fit_transform(y_train_1)
y_test_1 = label_encoder.transform(y_test_1)
```

```python
# Pipeline для кодирования категориальных признаков
ohe_pipe_1 = Pipeline(
    [
        ('simpleImputer_ohe', 
         SimpleImputer(missing_values=np.nan, 
                       strategy='most_frequent')), 
        ('ohe', 
         OneHotEncoder(drop='first', 
                       handle_unknown='error', 
                       sparse=False))
    ]
)
```

```python
#Pipeline для кодирования категориальных ранговых признаков
ord_pipe_1 = Pipeline(
    [('simpleImputer_before_ord', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
     ('ord',  OrdinalEncoder(
                categories=[
                    ['junior', 'middle', 'senior'], 
                    ['low', 'medium', 'high'],
                ], 
                handle_unknown='use_encoded_value', unknown_value=np.nan
            )
        ),
     ('simpleImputer_after_ord', SimpleImputer(missing_values=np.nan, strategy='most_frequent'))
    ]
)
```

```python
# Pipeline для подготовки данных
data_preprocessor_1 = ColumnTransformer(
    [
        ('ohe', ohe_pipe, ohe_col_1),
        ('ord', ord_pipe, ord_col_1),
        ('num', StandardScaler(), num_col_1)
    ], 
    remainder='passthrough'
)
```

```python
# Итоговый Pipeline: подготовка данных и модель
pipe_final_1 = Pipeline([
    ('preprocessor', data_preprocessor_1),
    ('models', DecisionTreeClassifier(random_state=RANDOM_STATE))
])
```

```python
#Параметры для передачи GridSearchCV для поиска оптимальной модели 
#с разными Scaler и Encoder и перебором гиперпараметров

param_distributions_1 = [
    {
        'models': [KNeighborsClassifier()],
        'models__n_neighbors': range(1, 11),
        'preprocessor__num': [StandardScaler(), MinMaxScaler()]   
    },
    {
        'models': [DecisionTreeClassifier(random_state=RANDOM_STATE)],
        'models__max_depth': range(2, 11),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
    },
    {
        'models': [SVC(random_state=RANDOM_STATE, probability=True)],
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  

    },
    {
        'models': [LogisticRegression(random_state=RANDOM_STATE)],
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
    }
] 
```

### Обучение моделей

```python
#поиск оптимальной модели с перебором гиперпараметров
grid_1 = GridSearchCV(
    pipe_final_1, 
    param_grid=param_distributions_1, 
    cv=5, 
    scoring='roc_auc', 
    n_jobs=-1, 
)
```

```python
grid_1.fit(X_train_1, y_train_1);
```

```python
print('Метрика ROC-AUC для лучшей модели:\n', round(grid_1.best_score_, 2)) 
print('\nЛучшая модель и её параметры:\n\n', grid_1.best_params_) 
```

```python
y_pred_1 = grid_1.predict_proba(X_test_1)
print(f'Метрика ROC-AUC на тестовой выборке: {round(roc_auc_score(y_test_1, y_pred_1[:, 1]), 2)}')
```

### Промежуточные выводы

```python
#поиск важных признаков
X_encoded_1 = grid_1.best_estimator_['preprocessor'].fit_transform(X_train_1)

model_1 = grid_1.best_estimator_['models'].fit(X_encoded_1, y_train_1)
explainer_1 = shap.PermutationExplainer(model_1, X_encoded_1)


labels_1 = grid_1.best_estimator_[:-1].get_feature_names_out()

plt.title('Оценка важности признаков')
plt.ylabel('Признаки')
shap.summary_plot(X_encoded_1, plot_type='bar', feature_names=labels_1) 
```

Лучшей моделью оказалась модель логистической регресии с метрикой ROC_AUC = 0.93. На тестовой выборке метрика составила 0.92, что удовлетворяет условиям задачи, где критерием успеха является ROC_AUC >= 0.91.
